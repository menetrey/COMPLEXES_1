## I) Construction des nombres complexes

### 1) Définition
On appelle l’ensemble des nombre complexes, noté $\mathbb{C}$, l’ensemble des nombres $z$ de la forme :  
$z = a + ib$ avec $(a, b) \in \mathbb{R}^2$ et $i^2 = −1$.  
  
  
- Le nombre réel $a$ s’appelle la partie réelle de $z$ notée : $Re(z)$.  
- Le nombre réel $b$ s’appelle la partie imaginaire de $z$ noté : $Im(z)$.   

Cette forme $z = a + ib$ est appelée forme algébrique.  

*remarques:*  
1) Tout nombre réel appartient à $\mathbb{C}$ (cas où $b = 0$).  
2) Si $a = 0$ on dit que $z$ est un imaginaire pur.  
3) Pour qu’un nombre complexe soit nul, il faut et il suffit que sa partie réelle et sa partie imaginaire soient nulles :
$a + ib = 0 \iff a = 0$ et $b = 0$
  

### 2) Représentation des nombres complexes

**Théorème:**   
A tout nombre complexe $z = a + ib$, on peut faire correspondre un point $M(a; b)$ dans un plan orthonormal $(O, \overrightarrow{u} , \overrightarrow{v} )$.  

On dit que $z$ est l’affixe de $M$. On écrit alors $M(z)$.  
  
  La réciproque est vraie.  
    
On peut représenter alors le nombre complexe $z = a + ib$.  
#### a) Module de z  

On appelle module de $z$ la distance OM, c’est-à-dire la quantité notée $|z|$ telle que :
$|z| =\sqrt{a^2+b^2}$

<img src="Complexe1.gif" width="200"/>

Si $z \in \mathbb{R}$, on a $z = a$ et $|z| = \sqrt{a^2} = |a|$ qui n’est autre que la valeur absolue du réel a.  

#### b) Argument de z  
Et pour $z \neq 0$, on appelle argument de $z$, noté $arg(z)$, toute mesure $\theta$ de l’angle
$(\overrightarrow{u};\overrightarrow{OM})$ telle que :  
$\Bigg\{ \begin{matrix}
cos(\theta)=\frac{a}{|z|}\\
sin(\theta)=\frac{b}{|z|}
\end{matrix} $ 
avec $\theta=arg(z)$   $ [2\pi]$

### 3) Opérations avec les complexes
#### a) Addition et multiplication
Dans l’ensemble des nombres complexes on définit deux opérations:  
• L’addition:  
si $z = a + ib$ et $z′ = a′ + ib′$ alors $z + z′ = (a + a′) + i(b + b′)$   
  
• La multiplication:  
si $z = a + ib$ et $z′ = a′ + ib′$ alors $z × z′ = (aa′ − bb′) + i(ab′ + a′b)$  
  

*Exemple:*  
Soit les opérations suivantes:  
$z_1 = 4 + 7i − (2 + 4i) = 4 + 7i − 2 − 4i = 2 + 3i$  
$z_2 = (2 + i)(3 − 2i)$  
$z_3 = (4 − 3i)^2 $  

#### b) Vecteur
Il est possible de représenter l'affixe d'un vecteur dans le plan.   
On peut retrouver l'affixe du vecteur $\overrightarrow{AB}$ par $z_{\overrightarrow{AB}}=z_B-z_A$ .  

*Exemple:*  
Représenter le vecteur image de $3+i$.

# II) Equation du second degré
## 1) Résolution

Si la résolution des équations du second degré dépendait du discriminant $\Delta$ c'est pour que $\sqrt{b^2-4ac}$ puisse exister. Or avec les nombres complexes, il est maintenant possible de lever cette difficulté.  
 
Donc en résumé:  
- Si $\Delta \gt 0$, l'équation admet deux solutions **réelles**.  
- Si $\Delta = 0$, l'équation admet une solution double **réelles**.   
- Si $\Delta \lt 0$, l'équation admet des solutions **complexes**.  

Soit $az^2+bz+c=0$ une équation à résoudre dans $\mathbb{C}$ avec $a,b,c  \in \mathbb{R}$ et $a \neq 0$.    

Si $\Delta \gt 0$, Il existe deux solutions **réelles**:   

$x_1=\frac{-b-\sqrt{\Delta}}{2a}$ et $x_2=\frac{-b+\sqrt{\Delta}}{2a}$.  

Si $\Delta = 0$, il existe une solution double réelle: $x_0=\frac{-b}{2a}$.  

Si $\Delta \lt 0$, il existe deux solutions **complexes**:      

$z_1=\frac{-b-i\sqrt{\Delta}}{2a}$ et  $z_2=\frac{-b+i\sqrt{\Delta}}{2a}$.
